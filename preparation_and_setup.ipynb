## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data

```python
input_path = '../../DATA/AIWH-State-and-territory-community-mental-health-care-2018-19/generated_data/'
input_file = 'transformed_data.csv'
```

```python
df = pd.read_csv(input_path + input_file, sep=',', encoding='utf-8')
df
```

As you can see in the above table, the generated data contains the rate of "Mental health service contacts" in state and territory areas between 2018 and 2019.

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = ['NSW', 'VIC', 'QLD', 'WA', 'SA', 'TAS', 'ACT', 'NT']
rate = list(df[['NSW', 'VIC', 'QLD', 'WA', 'SA', 'TAS', 'ACT', 'NT']].iloc[0])

x_pos = [i for i, _ in enumerate(x)]
plt.figure(figsize=(20,11))
plt.bar(x_pos, rate, color='orange')
plt.xlabel("State and territory", fontsize=18)
plt.ylabel("per 1,000 population", fontsize=18)
plt.title("Mental health care service contacts rate between 2018 and 2019", fontsize=28)

for index, value in enumerate(rate):
    plt.text(index - 0.24, value - 30, str(round(value, 2)), fontsize=18)

plt.xticks(x_pos, x, fontsize=16)

plt.savefig('img/Mental health care service contacts rate between 2018 and 2019.pdf')

plt.show()
```

The data described in the above figure shows that the rate of mental health care service contacts across Australia. The x-axis represents different states and territories in Australia and the y-axis represents the rate of mental health care service access per 1,000 population.

According to the figure, the ACT had the highest rate of mental health services access across all states and territories (750 per 1,000 population)  followed by SA (406 per 1,000 population) and QLD (401 per 1,000 population). At the other end of the scale, TAS reported the lowest rate (212 per 1,000 population)
